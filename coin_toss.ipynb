Simulate a coin toss, dice rolls, or random walks.

```python
import numpy as np
```

```python
p_t = 0.5
p_h = 0.5
```

```python
outcomes = ["Heads", "Tails"]
```

```python
results = np.random.choice(outcomes, 10000, p=[p_t, p_h])
```

```python
counts = {outcome: np.sum(results == outcome) for outcome in outcomes}
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(counts.keys(), counts.values(), color=['blue', 'orange'])
plt.xlabel('Outcomes')
plt.ylabel('Frequency')
plt.title('Frequency Distribution of coin toss outcomes')
```
